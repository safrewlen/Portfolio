```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import locale
locale.setlocale(locale.LC_NUMERIC, '')
import os
```

```python
import os
import numpy as np
array_with_dfs = []
for subdir, dirs, files in os.walk(r'C:\Users\safre\Desktop\diplom\pvloop\data\Occlusions'):
    for file in files:
        filepath = os.path.join(subdir, file)
#         print(file, subdir)
        if 'saline' in file:
            continue
        a = pd.read_csv(filepath, sep='\t', engine = 'python', decimal = ',', index_col=0)
        a.replace(' ', np.nan, inplace=True)
        a.dropna(inplace=True)
        a = a[:-5]
#         try:
        a = a.applymap(locale.atof)
#         except:
#             print(file, subdir)
#             continue
        a.reset_index(drop=True, inplace=True)
        a['group_type'] = subdir.split('\\')[-1]
        a['rat_no'] = file[:-4] 
        a['CE'] = a['SW (mmHg*µL)']/a['PVA (mmHg*µL)']
        a['IVG'] = a['dP/dt max (mmHg/s)']/a['P@dP/dt max (mmHg)']
        a = a.drop(['Vmax (µL)', 'Vmin (µL)', 'Pmax (mmHg)', 'Pmin (mmHg)', 'Pmean (mmHg)', 'Pdev (mmHg)',
                                 'dP/dt min (mmHg/s)', 'dV/dt max (µL/s)', 'dV/dt min (µL/s)', 'P@dV/dt max (mmHg)', 
                                'P@dP/dt max (mmHg)', 'V@dP/dt max (µL)', 'V@dP/dt min (µL)', 'Tau (ms)'], axis=1)
        array_with_dfs.append(a)
        #         print(subdir.split('\\')[-1])
#         break

array_with_dfs[0]
result_df = pd.concat(array_with_dfs)
result_df.head()
# result_df[result_df['rat_no'] == 'rat_10k_baseline_1']#['SW (mmHg*µL)']
```

```python
df_ce_means = pd.DataFrame(result_df.groupby(['group_type', 'rat_no'])['CE'].mean())
reshape = df_ce_means.pivot_table(index='rat_no', columns='group_type')
reshape = reshape[reshape.rat_no != 'rat_17_occlusion']
reshape

# reshape = reshape.as_matrix(columns=None)
# plt.boxplot(x = reshape)
# plt.title('all baselines dox', color='green')
# plt.ylabel('SW')
# plt.show()
# reshape.boxplot()
```
